7.Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df['income_binary'] = df['income'].map({'<=50K': 0, '>50K': 1})
display(df[['income', 'income_binary']].head())
```

8.Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [0, 30, 60, float('inf')]
labels = ['Youth', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
display(df[['age', 'age_group']].head())
```

9.Rename the 'educational-num' column to 'education_num'.

```python
df = df.rename(columns={'educational-num': 'education_num'})
display(df.head())
```

10.Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
df['workclass'] = df['workclass'].replace('?', 'Unknown')
df['occupation'] = df['occupation'].replace('?', 'Unknown')
display(df.head())
```

11.Calculate the average age for each workclass.

```python
average_age_by_workclass = df.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

12.Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = df.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

13.Get the count of each marital-status category.

```python
marital_status_counts = df['marital-status'].value_counts()
display(marital_status_counts)
```

14.Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_per_week_by_education = df.groupby('education')['hours-per-week'].mean()
display(mean_hours_per_week_by_education)
```

15.Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = df.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

16.Get the unique values and their counts for the 'native-country' column

21.Calculate the median income by education level and gender.

```python
df['income_binary'] = df['income'].map({'<=50K': 0, '>50K': 1})
median_income_by_education_gender = df.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

22.Find the top 3 occupations with the highest average capital-gain.

## Chart 1: Age Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Observation from the 'Distribution of Age' chart:
 The age distribution shows a peak around 30-40 years, indicating a younger working population, and gradually declines with increasing age.

## Chart 2: Income Distribution by Gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='gender', hue='income')
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Observation from the 'Income Distribution by Gender' chart:**

The chart shows that for both males and females, the majority of individuals have an income of `<=50K`. However, there is a clear disparity: a significantly higher number of males have an income of `>50K` compared to females. This suggests a potential gender gap in higher income brackets within the dataset.

## Chart 3: Workclass Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='workclass', hue='workclass', palette='viridis', legend=False)
plt.title('Distribution of Workclass')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

**Observation from the 'Distribution of Workclass' chart:**

The bar chart clearly shows that 'Private' is the predominant workclass, accounting for a significantly larger number of individuals compared to other categories. 'Self-emp-not-inc', 'Local-gov', and 'State-gov' are the next most frequent workclasses, while categories like 'Federal-gov', 'Self-emp-inc', 'Unknown', 'Without-pay', and 'Never-worked' represent smaller portions of the dataset. This indicates that the majority of the population in this dataset is employed in the private sector.

## Chart 4: Hours Per Week Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['hours-per-week'], bins=20, kde=True)
plt.title('Distribution of Hours Per Week')
plt.xlabel('Hours Per Week')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Observation from the 'Distribution of Hours Per Week' chart:**

The histogram shows a strong peak around 40 hours per week, indicating that a significant portion of individuals in the dataset work a standard full-time work week. There are also smaller peaks at lower and higher hours, but 40 hours is clearly the most common working duration. The distribution appears right-skewed, meaning there are fewer individuals working very long hours compared to those working fewer than 40 hours, but a substantial number still work above 40 hours.

## Chart 5: Education Level vs. High Income Proportion

```python
income_proportion_by_education = df.groupby('education')['income_binary'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x=income_proportion_by_education.index, y=income_proportion_by_education.values, hue=income_proportion_by_education.index, palette='viridis', legend=False)
plt.title('Proportion of High Income (>50K) by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion of >50K Income')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

**Observation from the 'Proportion of High Income (>50K) by Education Level' chart:**

The bar chart illustrates a strong positive correlation between education level and the proportion of individuals earning more than >50K. 'Prof-school' and 'Doctorate' consistently show the highest proportions of high-income earners, followed closely by 'Masters' and 'Bachelors'. Conversely, individuals with lower education levels such as '1st-4th', 'Preschool', and '5th-6th' have the lowest proportions of high income, suggesting that higher educational attainment is a significant factor in achieving higher income levels within this dataset.
